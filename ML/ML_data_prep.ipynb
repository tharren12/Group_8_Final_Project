```python
# Import our dependencies
import pandas as pd
from path import Path
from sklearn.preprocessing import LabelEncoder

# Read in our housing data

file_path = Path("../Resources/City_Zhvi_AllHomes.csv")
all_homes_df = pd.read_csv(file_path)

# Confirm dataload
all_homes_df
```

```python
all_homes_df.drop('RegionID', axis=1, inplace=True)
all_homes_df
```

```python
#Check for nulls
for column in all_homes_df.columns:
    print(f"column {column} has {all_homes_df[column].isnull().sum()} null values")
```

```python
# Remove all rows with null values since rows with null values are less than 15% of total dataset
all_homes_cleansed_df = all_homes_df.dropna(how='any',axis=0) 
```

```python
# Confirm none of the top 50 ranked cities were lost in the above step
all_homes_cleansed_df.head(50)
```

```python
le = LabelEncoder()
all_homes_cleansed_df['RegionName'] = le.fit_transform(all_homes_cleansed_df['RegionName'])
all_homes_cleansed_df['RegionType'] = le.fit_transform(all_homes_cleansed_df['RegionType'])
all_homes_cleansed_df['State'] = le.fit_transform(all_homes_cleansed_df['State'])
all_homes_cleansed_df['StateName'] = le.fit_transform(all_homes_cleansed_df['StateName'])
all_homes_cleansed_df['CountyName'] = le.fit_transform(all_homes_cleansed_df['CountyName'])
```
